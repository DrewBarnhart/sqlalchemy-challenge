```python
%matplotlib inline
from matplotlib import style
style.use('fivethirtyeight')
import matplotlib.pyplot as plt
```

```python
import numpy as np
import pandas as pd
import datetime as dt
```

# Reflect Tables into SQLAlchemy ORM

```python
# Python SQL toolkit and Object Relational Mapper
import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, func
```

```python
# create engine to hawaii.sqlite
engine = create_engine("sqlite:///Resources/hawaii.sqlite")
```

```python
pd.read_sql("Select * from measurement",con=engine)
```

```python
# reflect an existing database into a new model

Base = automap_base()

Base.prepare(autoload_with=engine)

# reflect the tables
```

```python
# View all of the classes that automap found
Base.classes.keys()
```

```python
# Save references to each table
Measurement= Base.classes.measurement
Station= Base.classes.station
```

```python
# Create our session (link) from Python to the DB
#session allows python to allow SQL to run within the database (ie clocking in and out)

session = Session(engine)
```

# Exploratory Precipitation Analysis

```python
# Find the most recent date in the data set.
```

```python
for row in session.query(Measurement).all():
    print(row.station, row.date, row.prcp, row.tobs)
```

```python
# session.query(Measurement.date).all()
# to show all

session.query(func.max(Measurement.date)).all()
```

```python
# Design a query to retrieve the last 12 months of precipitation data and plot the results. 
# Starting from the most recent data point in the database. 


# Calculate the date one year from the last date in data set.
last12months = dt.date(2017,8,23)-dt.timedelta(days=365)


# Perform a query to retrieve the data and precipitation scores

dandp = session.query(Measurement.date,Measurement.prcp).filter(Measurement.date>= last12months).all()


# Save the query results as a Pandas DataFrame and set the index to the date column

dandp_df = pd.DataFrame(dandp)


# Sort the dataframe by date

dandp_df = dandp_df.set_index("date")
dandp_df
dandp_df= dandp_df.sort_index(ascending = True)
dandp_df

# Use Pandas Plotting with Matplotlib to plot the data

dandp_df.plot(rot=53)
plt.ylabel("Inches")
plt.title("last 12 months of precipitation data")
```

```python
# Use Pandas to calcualte the summary statistics for the precipitation data
dandp_df.describe()
```

```python
#used to provide information regarding data frame
dandp_df.info()
```

# Exploratory Station Analysis

```python
# Design a query to calculate the total number stations in the dataset
session.query(func.count(Station.station)).all()
```

```python
# Design a query to find the most active stations (i.e. what stations have the most rows?)
# List the stations and the counts in descending order.
session.query(Measurement.station,func.count(Measurement.prcp)).group_by(Measurement.station
                                                                        ).order_by(func.count(Measurement.prcp).desc()).all()
```

```python
# withot groupby 
# session.query(Measurement.station,func.count(Measurement.prcp)).order_by(func.count(Measurement.prcp).desc()).all()
```

```python
# Using the most active station id from the previous query, calculate the lowest, highest, and average temperature.
session.query(func.min(Measurement.tobs),
              func.max(Measurement.tobs),
              func.avg(Measurement.tobs)
              ).filter(Measurement.station=="USC00519281").first()
```

```python
# Using the most active station id
# Query the last 12 months of temperature observation data for this station and plot the results as a histogram

stationdata = session.query(Measurement.tobs).filter((Measurement.date>= last12months
                                                     ) & (Measurement.station=="USC00519281")).all()

stationdatadf = pd.DataFrame(stationdata)

stationdatadf.plot.hist(bins=12)
plt.xlabel("Temperature")
plt.title("USC00519281 Frequency of Temperature Over 12 Months")

plt.show()
```

# Close session

```python
# Close Session
session.close()
```
